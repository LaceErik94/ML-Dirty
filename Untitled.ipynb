```python
# Identify categorical variables that can be transformed directly
transformable_categorical = [col for col in categorical_cols if df[col].nunique() <= 10]
print(f"Categorical Variables Suitable for Direct Transformation: {transformable_categorical}\n")
```

```python
outlier_counts = {}
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outlier_counts[col] = ((df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))).sum()

print(f"Outlier Counts in Numerical Variables:\n{outlier_counts}\n")
```

```python
# Define a function to clean and convert
def safe_convert(val):
    if isinstance(val, str):
        val = val.lower().strip()  # Normalize case and strip spaces
        if val in ["zero", "none", "?"]:  # Replace known non-numeric values
            return 0.0  # Or use np.nan if you prefer missing values
    try:
        return float(val)  # Convert valid numbers
    except ValueError:
        return np.nan  # Handle unexpected cases

# Apply the conversion to all object columns
df[df.select_dtypes(include='object').columns] = df.select_dtypes(include='object').applymap(safe_convert)

# Verify conversion
print(df.dtypes)  # Ensure object columns are now float
```

```python
# Compute the Spearman correlation between numerical and converted categorical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns  # All numerical columns
converted_categorical_cols = df.select_dtypes(include=['float64']).columns.difference(numerical_cols)  # Only converted ones

# Calculate Spearman correlation
spearman_corr = df[numerical_cols].corrwith(df[converted_categorical_cols], method='spearman')

# Display results
print(spearman_corr)
```

```python
interest_variables = [
    "word_freq_free", "word_freq_money", "word_freq_credit", "word_freq_business",
    "word_freq_email", "word_freq_you", "word_freq_your", "word_freq_000",
    "char_freq_!", "char_freq_$", "char_freq_#",
    "capital_run_length_average", "capital_run_length_longest", "capital_run_length_total"
]

spam_correlation = df[interest_variables + ["is_spam"]].corr()["is_spam"].sort_values(ascending=False)
print(spam_correlation)
```
